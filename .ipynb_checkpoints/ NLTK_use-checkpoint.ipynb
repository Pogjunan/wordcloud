```python
filename = 'riss_bigdata.csv'
files_df = pd.read_csv(filename)
```

```python
all_title = files_df['제목']
all_title
```

```python
stopWords = set(stopwords.words('english'))
lemma = WordNetLemmatizer() #표제어 추출함수 running-> run
```

```python
words = []

for title in all_title:
    EnWords = re.sub(r"[^a-zA-Z]+", " ",str(title))
    EnWordsToken = word_tokenize(EnWords.lower())
    EnWordsTokenStop = [w for w in EnWordsToken if w not in stopWords]
    EnWordsTokenStopLemma = [lemma.lemmatize(w) for w in EnWordsTokenStop]
    words.append(EnWordsTokenStopLemma)
```

```python
#print(words)
words2 = list(reduce(lambda x,y : x+y,words))
print(words2)
```

```python
count =Counter(words2)
count
```

```python
word_count = dict()

for tag,counts in count.most_common(50):
    if(len(str(tag))>1):
        word_count[tag] = counts
        print("%s :%d" % (tag,counts))
```

```python
del word_count['big']
del word_count['data']
```

```python
sorted_Keys = sorted(word_count , key = word_count.get,reverse=True)
sorted_Values = sorted(word_count.values() , reverse=True)
```

```python
# 단어 비녿 히스토그램
```

```python
plt.figure(figsize=(12,5))
plt.xlabel("word")
plt.ylabel("count")
plt.grid(True)

plt.bar(range(len(word_count)),sorted_Values, align='center')
plt.xticks(range(len(word_count)),list(sorted_Keys),rotation=85)
plt.show()
```

```python
files_df['doc_count'] = 0
summary_year = files_df.groupby('출판일',as_index=False)['doc_count'].count()
summary_year
```

```python
plt.figure(figsize=(12,5))
plt.xlabel("year")
plt.ylabel("doc-count")
plt.grid(True)

plt.bar(range(len(summary_year)),summary_year['doc_count'])
plt.xticks(range(len(summary_year)),[text for text in summary_year['출판일']])
plt.show()
```

```python
stopwords=set(STOPWORDS)
wc = WordCloud(background_color = 'ivory', stopwords= stopwords, width=800, height=400)
cloud =wc.generate_from_frequencies(word_count)

plt.figure(figsize=(8,8))
plt.imshow(cloud)
plt.axis('off')
plt.show()
```
